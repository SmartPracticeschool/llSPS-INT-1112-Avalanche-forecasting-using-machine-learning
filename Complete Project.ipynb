# 6.One Hot Encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer([("on",OneHotEncoder(),[1])],remainder="passthrough")
x=ct.fit_transform(x)
x
```

```python
x=x[:,1:]
x
```

# 8.Splitting Data Into Train and Test

Firstly,We have to split the data into Train and Test to perform Feature Scaling

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

# 7.Feature Scaling

To perform feature scaling we must perform any one type of Encoding(Label or One Hot)

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# Model Building

# Training And Testing The Model

We used Decision Tree Classifier to Training and Testing the model

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=0,criterion='gini')
dt.fit(x_train,y_train)
```

```python
y_predict=dt.predict(x_test)
y_predict
```

```python
y_test
```

# Model Evaluation

We can predict the output by giving some values

```python
dt.predict([[1.0,0.0,54,24,15,19]])
```

```python
accuracy=accuracy_score(y_test,y_predict)
print("Accuracy of the model:", accuracy)
```

So,After using Decision Tree Classifier we got 90% Accuracy
