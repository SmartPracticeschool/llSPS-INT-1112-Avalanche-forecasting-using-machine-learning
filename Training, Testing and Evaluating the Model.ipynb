# 6.One Hot Encoding

# 7.Feature Scaling

To perform feature scaling we must perform any one type of Encoding(Label or One Hot)

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# Training And Testing The Model

We used Decision Tree Classifier to Training and Testing the model

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=0,criterion='gini')
dt.fit(x_train,y_train)
```
